```
import pandas as pd
import os
import numpy as np


root = './gdrive/My Drive/thesis'
df = pd.read_csv(os.path.join(root, 'Database completo .csv'), header=0, index_col=0, delimiter=';')
result = pd.read_csv(os.path.join(root, 'miei_result.csv'), header=0, index_col=0)
df = df.loc[result.index.values]
print(df)
new_result = result.assign(Diagnosi=df['DGN1'])
print(new_result)
```

```
import math
nuovi_normali = pd.read_csv(os.path.join(root, 'nuovi_result.csv'), header=0, index_col=0)
def map_index(s):
  n = int(s.split('-')[-1])
  return 'N'+str(int(math.ceil(n/2)))

normali_index = nuovi_normali.index.values
nuovi_index = list(map(map_index, normali_index))
print(nuovi_index)
nuovi_normali = nuovi_normali.set_index([nuovi_index])
print(nuovi_normali)
```

```
nuovi_normali = nuovi_normali.assign(Diagnosi=['NORMALI' for i in range(len(nuovi_normali))])
print(nuovi_normali)
```

```
new_result = new_result.append(nuovi_normali)
print(new_result)
```

```
new_result.to_csv(os.path.join(root, 'nuovo_dataset_compl.csv'))
```
